```python
avocado_dataset.sort_values(by=['Date'], inplace=True, ascending=True)
```

```python
# Average price of Conventional Avocados over time

mask = avocado_dataset['type']== 'conventional'
plt.rc('figure', titlesize=20)
fig = plt.figure(figsize = (27, 12))
fig.suptitle('Average Price of Conventional Avocados Over Time', fontsize=25)
ax = fig.add_subplot(111)

dates = avocado_dataset[mask]['Date'].tolist()
avgPrices = avocado_dataset[mask]['AveragePrice'].tolist()

plt.scatter( dates,avgPrices, c=avgPrices, cmap='plasma')
ax.set_xlabel('Date',fontsize = 15,)
ax.set_ylabel('Average Price (USD)', fontsize = 15)
ax.set_xlim()
ax.tick_params(labelrotation=90)

plt.show()
```

```python
av2_dataset=avocado_dataset.copy()
```

```python
# Creating Datframe for numeric features
av2_dataset.drop(['Date','type','region','year'], axis=1, inplace=True)
```

```python
plt.figure(figsize=(20,25))
plotnumber =1
for column in av2_dataset:
    if plotnumber <=9:
        ax = plt.subplot(3,3,plotnumber)
        sns.distplot(av2_dataset[column], color='r',hist=False,kde_kws={"shade": True})
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.show()
```

```python
sns.set_palette('gist_rainbow_r')
plt.figure(figsize=(20,20), facecolor='white')
plotnumber =1
for column in av2_dataset:
    if plotnumber <=9:
        ax = plt.subplot(3,3,plotnumber)
        sns.violinplot(avocado_dataset[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.show()
```
